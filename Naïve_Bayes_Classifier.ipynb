```python
#First, we will create a frequency table using each attribute of the dataset.
prior = data.groupby('class').size().div(len(data)) 
likelihood = {}
for key in name:
    likelihood[key]=data.groupby(["class", key])[key].count()
print(likelihood) 
```

```python
#will generate a likelihood table.
count_yes=0
count_no =0
likelihood2 = {}
likelihood3 = {}
likelihood4 = {}
likelihood4_to = {}
lables={}
lables2={}
for i in range(len(data)):
    if (data["class"][i]=="yes"):
        count_yes+=1
    else:
        count_no+=1
P_Y=count_yes/len(data)
P_N=count_no/len(data)
pro_yes=1
pro_no=1
for key in name:
    likelihood2[key]=data.groupby(["class", key])[key].count()/len(data)
for key in name:
    likelihood3[key]={'no':likelihood[key]["no"]/likelihood[key]["no"].sum(),'yes':likelihood[key]["yes"]/likelihood[key]["yes"].sum()}

```

```python

data=pd.read_csv ('test.csv')
for keys,values in labels3.items():
    x=max(len(likelihood2[keys]["no"]),(len(likelihood2[keys]["yes"])))
    if keys!="class":
        for j in range(x):
            if(values[j]in likelihood[keys]["no"] and values[j]in likelihood[keys]["yes"]):   
                likelihood4_no[values[j]]=round((likelihood[keys]["no"][values[j]]+likelihood[keys]["yes"][values[j]])/len(data),6)
            if(values[j]in likelihood[keys]["no"] and not values[j] in likelihood[keys]["yes"]):
                likelihood4_no[values[j]]=round((likelihood[keys]["no"][values[j]])/len(data),6)
            if(values[j]in likelihood[keys]["yes"] and not values[j] in likelihood[keys]["no"]):
                likelihood4_no[values[j]]=round((likelihood[keys]["yes"][values[j]])/len(data),6)
        likelihood4_to[keys]=(likelihood4_no)
        likelihood4_no={}
    else:
        break;
print("/______________________________________(x|class)______________________________________/")
for keys,values in likelihood3.items():
    print(keys,values)
print("/______________________________________(x)______________________________________/")
for keys,values in likelihood4_to.items():
    if(keys!="class"):
        print(keys,values)
                         
```
